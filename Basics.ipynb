```python
# share_man
recent_grads['ShareMen'] = recent_grads['Men'] / recent_grads['Total']
recent_grads['ShareMen'].head(10)
```

```python
ax1=recent_grads.plot(x='Sample_size', y='Median', kind='scatter', xlim =(0, 5000))
ax1.set_title('Fig1: Sample_size vs. Median')
```

The plot tells that data has a high variance near the 0 points but quickly starts to diminish as Full_time becomes higher.  Most of the density of students can be found in unpopular courses. It can be seen that mostly students from unpopular courses are getting paid more than the popular courses.

Therefore, it can clearly be said that students from popular courses do not tend to get paid more.

```python
ax2=recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_title('Fig2: Sample size vs. Unemployment Rate')
```

Smaller the sample size, bigger the Unemployment_rate span, as expected.
There is a lot of variation in unemployment rates among majors with small sample sizes. Yet again the small sample sizes may affect the representativeness of the relationship plotted here.

In addition, from reviewing some rows of the dataframe, there is a noticeable difference between the sample size and the number of graduates for whom there is data on whether they are employed/unemployed.

For example for Petroleum Engineering (rank 1) there is a sample size of 36 and a total of (1976+37) employed and unemployed.

This suggests that sample size is not readily comparable to other statistics collected, other than median wage.

```python
ax3=recent_grads.plot(x='Full_time', y = 'Median', kind="scatter")
ax3.set_title('Fig3: Median salary vs. Full time employment')
```

There is not a noticeable correlation between the number of graduates per major employed full-time and the median wage. If there was to be a relationship it would be positive i.e. more full-time employees leads to a higher median wage.

But as noted above the median wage figures are based off smaller unweighted samples that may not represent the wider population of graduates with each major.

# Visualizing the data: Bar plots
Using either df.plot(kind='bar') or df.plot.bar(x=labels, y= data for bars)

```python
#Compare the percentages of women from the first and last ten rows
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', title=('Share of women in the 10 courses with the highest median salary'))
```

```python
recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', title=('Share of women in the 10 courses with the lowest median salary'))
```

The courses with the highest median salaries have a lower share of female graduates than those with the lowest median salaries, which are majority female (i.e. over 50% of the graduates are women).

Now to calculate how large the difference is:

```python
#Calculating the average proportion of female graduates for the top and bottom 10 courses
top_10_female_share = recent_grads.loc[:9, "ShareWomen"].mean()
bottom_10_female_share = recent_grads[-10:]['ShareWomen'].mean()
```

```python
top_10 = "The 10 highest paying courses have an average female share of {:.2f}".format(top_10_female_share)
bottom_10 = "The 10 lowest paying courses have an average female share of {:.2f}".format(bottom_10_female_share)
print(top_10)
print(bottom_10)
```

```python
#Unemployment rate from top major and least major
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', title='Unemployment rate for the top 10 majors')
```

```python
recent_grads[-10:].plot.bar(x='Major', y='Unemployment_rate', title='Unemployment rate for the least 10 majors')
```

For this comparison it is less clear. The top 10 courses do tend to have a lower unemployment rate, apart from 2 exceptions: 'Nuclear Engineering' and 'Mining and Mineral Engineering'. Whilst for the bottom 10 courses 3-5 of the courses have higher unemployment rates.

This is can be analysed by looking at average unemployment rates.

```python
avg_unemploy=recent_grads['Unemployment_rate'].mean()
```

```python
top_avg_unemploy=recent_grads[:9]['Unemployment_rate'].mean()
least_avg_unemploy=recent_grads[-10:]['Unemployment_rate'].mean()
```

```python
mean_all='The average unemployment rate across all majors is {:.2f}'.format(avg_unemploy)
mean_top='The average unemployment rate across top majors is {:.2f}'.format(top_avg_unemploy)
mean_least='The average unemployment rate across least majors is {:.2f}'.format(least_avg_unemploy)
print(mean_all)
print(mean_top)  
print(mean_least)                                                                              
```

```python
top_ten_outliers=recent_grads[:10].loc[recent_grads['Unemployment_rate']>avg_unemploy]
top_ten_outliers['Difference from avg']=top_ten_outliers['Unemployment_rate']-avg_unemploy
top_ten_outliers.shape
```

```python
bottom_ten_outliers=recent_grads[-10:].loc[recent_grads['Unemployment_rate']>avg_unemploy]
bottom_ten_outliers['Difference from avg']=bottom_ten_outliers['Unemployment_rate']-avg_unemploy
bottom_ten_outliers
```

```python
top_ten_outliers.plot.bar(x='Major', y='Difference from avg', title='Majors in the top 10 with above average unemployment rates')
bottom_ten_outliers.plot.bar(x='Major', y='Difference from avg', title='Majors in the bottom 10 with above average unemployment rates')
```

So in the top 10, one course (Nuclear Engineering) is dragging up the average unemployment rate for the top 10 majors. In the bottom 10 Clinical Psychology has an above average unemployment rate, but the other 4 courses are also dragging up the unemployment rate.
